```python
#Plot graph for number of CT performed in January per date on an evening shift

import seaborn as sns
import matplotlib.pyplot as plt

x1eve = df1eve['verifieddate']  # Column for X-axis (e.g., timestamps)
y1eve = df1eve['count']         # Column for Y-axis (e.g., frequency)

# Create the bar plot
q1eve = sns.barplot(x=x1eve, y=y1eve)

# Set the title of the plot
q1eve.set(title='Evening Shift (0000-0800)')

# Set the X and Y labels
q1eve.set_xlabel("Date", fontsize=20)
q1eve.set_ylabel("Number of CT scans", fontsize=20)

# Add a horizontal line at y=40
q1eve.axhline(40)

# Rotate the X-axis tick labels by 90 degrees
plt.xticks(rotation=90)

# Add numerical values on top of each bar
for p in q1eve.patches:
    q1eve.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     xytext=(0, 5),
                     textcoords='offset points',
                     fontsize=10,
                     color='red')  # Change color to red

# Display the plot
plt.show()
```
